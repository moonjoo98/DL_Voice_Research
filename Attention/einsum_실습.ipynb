## 예제. 행렬의 내적

```python
# Dot Product. numpy로 행렬의 내적곱
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])

print(A)
print('------------------------')
print(B)
print('------------------------')
# np.einsum을 사용하여 행렬 곱셈 수행. 2*3 * 3*2 -> 2*2의 행렬
C = np.einsum('ik,kj->ij', A, B)

# 결과 출력
print(C)
```

##Unary Operation

```python
#Transpose. numpy로 2차원 행렬 전치
import numpy as np

A = np.array([[1,2,3], [4,5,6]])
print(A)
print('------------------------')
R = np.einsum("ij->ji", A)
print(R)
```

```python
#Transpose. torch로 2차원 행렬 전치
import torch

A = torch.randint(1, 10, (3,3)) #난수로 텐서 생성
print(A)
print('------------------------')
B = torch.einsum("ij->ji",A)
print(B)
```

```python
#diag, trace. torch로 대각 행렬과 합 구하기.
import torch

A = torch.randint(1, 10, (3,3))
diag = torch.einsum('ii->i', A)
trace = torch.einsum('ii->', A)
print(A) # 3x3 행렬
print('------------------------')
print(diag) # 대각행렬 구하기
print('------------------------')
print(trace) # 대각행렬의 합
```

## Summation

```python
# 2-D Matrix Summation
A = torch.randint(1, 10, (3,3))
print(A)
B = torch.einsum("ij->",A)
print('------------------------')
print(B)
```

```python
# 2-D Matrix Column sum
A = torch.randint(1,10,(3,3))
print(A)
B = torch.einsum('ij->j',A)
print('------------------------')
print(B)
```

```python
# 2-D Matrix Row sum
A = torch.randint(1,10,(3,3))
print(A)
B = torch.einsum('ij->i', A)
print('------------------------')
print(B)
```

## Matrix Multiplication

```python
# Matrix-Vector multiplication
A = torch.randint(0,10,(2,3))
B = torch.randint(0,10,(3,))
print(A)
print('------------------------')
print(B)

C = torch.einsum('ij,j->i',A,B)
print('------------------------')
print(C)
```

```python
# Matrix-Matrix multiplication
A = torch.randint(0,10,(2,3))
B = torch.randint(0,10,(3,2))
print(A)
print('------------------------')
print(B)

C = torch.einsum('ij,jk->ik',A,B)
print('------------------------')
print(C)
```

```python
#Batched Matrix multiplication
i,j,k,l = 2,2,2,3

A = torch.randint(0,10,(i,j,k))
print(A.shape)
print('------------------------')
print(A)

B = torch.randint(0,10,(i,k,l))
print('------------------------')
print(B.shape)
print('------------------------')
print(B)

# tensor shape 212,223 -> 213
C = torch.einsum('ijk,ikl->ijl',A,B)
print('------------------------')
print(C)
```

```python
#Bilinear Transformation
i, j, k, l= 2,3,2,2
A = torch.randint(0,10, (i,k))
X = torch.randint(0,10, (j,k,l))
B = torch.randint(0,10, (i,l))

print(A.shape)
print('------------------------')
print(A)
print('------------------------')
print(X.shape)
print('------------------------')
print(X)
print('------------------------')
print(B.shape)
print('------------------------')
print(B)

D = torch.einsum('ik,jkl,il->ij',A,X,B)
print('------------------------')
print(D)
```

## Dot/Outer/Hadamard Product

```python
# Dot Product. 내적
A = torch.arange(0,4)
B = torch.arange(2,6)
print(A)
print('------------------------')
print(B)

C = torch.einsum('i,i->',A,B)
print('------------------------')
print(C)
```

```python
# Outer Product. 외적
A = torch.arange(6)
B = torch.arange(6)
print(A)
print('------------------------')
print(B)

C = torch.einsum('i,j->ij',A,B)
print('------------------------')
print(C)
```

```python
# Hadamard Product.
A = torch.arange(6).resize(2,3)
B = torch.arange(6).resize(2,3)
print(A)
print('------------------------')
print(B)

C = torch.einsum('ij,ij->ij',A,B)
print('------------------------')
print(C)
```
